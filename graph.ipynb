```python
import numpy as np

import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

plt.figure(figsize=(7, 4))
plt.plot(x, tanh(x), label='tanh(x)', color='tab:blue')
plt.title('Đồ thị hàm Tanh')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.ylim(-1.05, 1.05)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

x = np.linspace(-6, 6, 1000)
y = sigmoid(x)

plt.figure(figsize=(7, 4))
plt.plot(x, y, label='sigmoid(x)', color='tab:blue')
plt.title('Đồ thị hàm Sigmoid')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.ylim(-0.05, 1.05)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
