```python
import numpy as np

import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

plt.figure(figsize=(7, 4))
plt.plot(x, tanh(x), label='tanh(x)', color='tab:blue')
plt.title('Đồ thị hàm Tanh')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.ylim(-1.05, 1.05)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

plt.figure(figsize=(7, 4))
plt.plot(x, gelu(x), label='gelu(x)', color='tab:blue')
plt.title('Đồ thị hàm GELU')
plt.xlabel('x')
plt.ylabel('gelu(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Tạo dữ liệu mẫu cho SVM
np.random.seed(42)
# Tạo hai nhóm dữ liệu
group1 = np.random.randn(20, 2) + [-2, -1]
group2 = np.random.randn(20, 2) + [2, 1]

# Vẽ sơ đồ SVM
plt.figure(figsize=(10, 8))

# Vẽ các điểm dữ liệu
plt.scatter(group1[:, 0], group1[:, 1], c='red', marker='o', s=100, alpha=0.7, label='Lớp -1')
plt.scatter(group2[:, 0], group2[:, 1], c='blue', marker='s', s=100, alpha=0.7, label='Lớp +1')

# Vẽ siêu phẳng (hyperplane) và các đường biên
x_hyperplane = np.linspace(-6, 6, 100)
y_hyperplane = 0.5 * x_hyperplane + 0.5

# Siêu phẳng chính
plt.plot(x_hyperplane, y_hyperplane, 'k-', linewidth=2, label='Siêu phẳng')

# Đường biên (margin)
margin = 1.2
plt.plot(x_hyperplane, y_hyperplane + margin, 'k--', linewidth=1, alpha=0.7, label='Biên trên')
plt.plot(x_hyperplane, y_hyperplane - margin, 'k--', linewidth=1, alpha=0.7, label='Biên dưới')

# Tô màu vùng margin
plt.fill_between(x_hyperplane, y_hyperplane - margin, y_hyperplane + margin, 
                 alpha=0.2, color='gray', label='Vùng biên (Margin)')

# Đánh dấu support vectors
support_vectors_1 = group1[np.abs(group1[:, 1] - (0.5 * group1[:, 0] + 0.5)) < 1.5][:3]
support_vectors_2 = group2[np.abs(group2[:, 1] - (0.5 * group2[:, 0] + 0.5)) < 1.5][:3]

plt.scatter(support_vectors_1[:, 0], support_vectors_1[:, 1], 
            s=200, facecolors='none', edgecolors='red', linewidth=3, label='Support Vectors')
plt.scatter(support_vectors_2[:, 0], support_vectors_2[:, 1], 
            s=200, facecolors='none', edgecolors='blue', linewidth=3)

plt.xlim(-6, 6)
plt.ylim(-4, 4)
plt.xlabel('Đặc trưng X₁', fontsize=12)
plt.ylabel('Đặc trưng X₂', fontsize=12)
plt.title('Sơ đồ biểu diễn SVM với Siêu phẳng', fontsize=14, fontweight='bold')
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

x = np.linspace(-6, 6, 1000)
y = sigmoid(x)

plt.figure(figsize=(7, 4))
plt.plot(x, y, label='sigmoid(x)', color='tab:blue')
plt.title('Đồ thị hàm Sigmoid')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.ylim(-0.05, 1.05)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
